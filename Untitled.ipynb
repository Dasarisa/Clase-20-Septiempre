```python
from numpy import linspace, sin, cos
x=linspace(0,10,100000)
y=sin(x)
z=cos(x)
plot(x,y)
plot(x,z)
show()
```

```python
from pylab import ylim, xlim, grid
x=linspace(0,10,100000)
y=sin(x)
plot(x,y)
xlim(-1.1, 1.1)
xlim(0., 10)
grid('on')
show()
```
